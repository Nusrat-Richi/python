```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/rahmanziaur/EDGEICT/refs/heads/main/HistoricalPrices.csv')

# # Load the dataset into a Pandas DataFrame
# df = pd.read_csv("HistoricalPrices.csv")

# Rename the column to remove an additional space
df = df.rename(columns = {' Open': 'Open', ' High': 'High', ' Low': 'Low', ' Close': 'Close'})

# Convert the date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Sort the dataset in the ascending order of date
df = df.sort_values(by = 'Date')

# Extract the date and close price columns
dates = df['Date']
closing_price = df['Close']

# Plot in Red colour
plt.plot(dates, closing_price, color='red')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/rahmanziaur/EDGEICT/refs/heads/main/HistoricalPrices.csv')

#rename columns name to remove white spaces at the beginning or end

df = df.rename(columns = {' Open': 'Open', ' High': 'High', ' Low': 'Low', ' Close': 'Close'})

range = df['Close'].max() - df['Close'].min()

df1 = df[['Date','Low']]

newlist=[]
for d, h in df1.iterrows():
  date = h['Date']
  Low = h['Low']

  if 42500 <= Low <= 43500:
    newlist.append([date, Low])

df2= pd.DataFrame(newlist, columns = ['Date', 'Low'])
df2.plot()


```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# creating the dataset
# data = {'C':20, 'C++':15, 'Java':30, 'Python':35}
x=df2['Date']
y=df2['Low']
fig = plt.figure(figsize = (10, 5))

# creating the bar plot
plt.bar(x, y, color ='blue',
        width = 0.3)

plt.xlabel("Date")
plt.ylabel("Lowest Value of the Stock")
plt.title("WSJ Stock Report for Lowest Value")
plt.show()
```
